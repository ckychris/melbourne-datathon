````python
import os

import pandas as pd
````

````python
data_dir = os.getenv('DATA_DIR')
print (data_dir)
````

````python
# Path to the data.  Forgive me for renaming everything!

fd_drugs = os.path.join(data_dir, "Lookups", "Drug_LookUp.txt")
print(fd_drugs)
````

````python
# Main dataframe, takes a bit of time to load.
drugs_frame = pd.read_csv(fd_drugs, sep = '\t', na_filter = False)
````

````python
drugs_frame.count()
````

````python
drugs_frame.head()
````

````python
# Looking for the number of unique element in each columns.

for col in drugs_frame.columns:
    print(col, len(drugs_frame[col].unique()))
````

````python
# variation to order the lists... couldn't work out how to do it on one line...
# Looking at the smaller sets to see what they might contain.
for col in ('PackSizeNumber', 'EthicalSubCategoryName', 'EthicalCategoryName'):
    u = drugs_frame[col].unique()
    u.sort()
    print(col, len(u), u)
````

Noticed that there were a few less `MasterProductFullName` instances to `MasterProductID` instances, so I found this 'look for duplicates' code in the internet (stackoverflow)

````python
pd.concat(g for _, g in drugs_frame.groupby("MasterProductFullName") if len(g) > 1)
````
